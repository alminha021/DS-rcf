```python
import random
import string

class Aluno:
    def __init__(self):
        self.nome = ''.join(random.choices(string.ascii_uppercase, k=12))
        self.matricula = ''.join(random.choices(string.digits, k=6))
        self.creditos = random.randint(0, 180)

def imprime(lista):
    for a in lista:
        print(a.nome, a.matricula, a.creditos, sep='\t')
    print()

def ordenaCreditos(A):
    n = len(A)
    count = [0] * 181

    for aluno in A:
        count[aluno.creditos] += 1
    for i in range(1, 181):
        count[i] += count[i - 1]

    B = [None] * n
    for i in range(n - 1, -1, -1):
        B[count[A[i].creditos] - 1] = A[i]
        count[A[i].creditos] -= 1

    for i in range(n):
        A[i] = B[i]
    print("Ordenado por credito: ")
    return A

def countingSortMatricula(A, exp, M):
    n = len(A)
    B = [None] * n
    count = [0] * M

    for aluno in A:
        index = (int(aluno.matricula) // exp) % 10
        count[index] += 1

    for i in range(1, M):
        count[i] += count[i - 1]

    i = n - 1
    while i >= 0:
        index = (int(A[i].matricula) // exp) % 10
        B[count[index] - 1] = A[i]
        count[index] -= 1
        i -= 1

    for i in range(n):
        A[i] = B[i]

def ordenaMatricula(A):
    max_element = max(int(aluno.matricula) for aluno in A)
    p = 1
    while max_element // p > 0:
        countingSortMatricula(A, p, 10)
        p *= 10
    print("Ordenado por matricula: ")
    return A

def counting_sort_nome(A, exp,M):
    n = len(A)
    B = [None] * n
    count = [0] * M
    #para nome , temos q usar o valor de cada letra na tabela unicode
    for aluno in A:
        index = ord(aluno.nome[exp]) - 65 #65 representa "A" na tabela unicode, e tambem a primeira letra do alfabeto, todas as seguintes sÃ£o em sequencia a 65
        count[index] += 1

    for i in range(1, M):
        count[i] += count[i - 1]

    i = n - 1
    while i >= 0:
        index = ord(A[i].nome[exp]) - 65
        B[count[index] - 1] = A[i]
        count[index] -= 1
        i -= 1

    for i in range(n):
        A[i] = B[i]

def ordenaNome(A):
    max_length = 12
    for p in range(max_length - 1, -1, -1):
        counting_sort_nome(A, p, 26)
    print("Ordenado por nome: ")
    return A

n = 20
alunos = [Aluno() for _ in range(n)]


imprime(alunos)
alunos1 = ordenaCreditos(alunos)
imprime(alunos1)
alunos2 = ordenaMatricula(alunos1)
imprime(alunos2)
alunos3 = ordenaNome(alunos2)
imprime(alunos3)
```
