```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

importando o dataset do excel para facilitar

```python
data = pd.read_excel("datateste.xlsx")
data.to_csv('dataset')
#df = pd.DataFrame(data, columns=['apple','alfa','boa'])
data.head()
```

Mapa de dipersao utilizando scatter do matplotlib, entre a coluna de valores para a apple e a coluna de valores para a alfabet. PEla disposição dos valores se nota uma correlaçao, a medida que um valor sobe, o outro valor tbm sobe.

```python
plt.scatter(data['apple'],data['alfa'])
plt.title('Grafico dispersao APPLE X Alfabet')
plt.show()
```

Mapa de dipersao utilizando scatter do matplotlib, entre a coluna de valores para a apple e a coluna de valores para a bank of america. PEla disposição dos valores nao se nota uma relaçao entre os valors, pouca relaçao, um valor nao influencia o outro

```python
plt.scatter(data['apple'],data['boa'])
plt.title('Grafico dispersao APPLE X B. of America')
plt.show()
```

utilizando o seaborn para traçar uma reta de regressao entre os 2 valores, a correlaçao eh semelhante a reta, onde a inclinaçao indica q o aumento de um valor, o outro tbm aumenta,

```python
sns.regplot(x=data['apple'], y = data['alfa'])
```

utilizando a biblioteca seaborn para traçar uma reta, no caso de regressao, mas da para perceber a pouca relaçao entre os valore, ja q a reta esta pouco inclinada

```python
sns.regplot(x=data['apple'], y = data['boa'])
```
