```python
import pandas as pd
from statistics import median
from math import isnan
from itertools import filterfalse
import statistics
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
escrita = [344,383,611,713,536]
leitura = [378,349,503,719,503]
```

Considerando que os dados representam uma amostra de um conjunto de dados maiores = seria uma amostra.

Se considerar que existem apenas esses 5 alunos, ele representa todo o conjunto da populaçao, seria uma população.

NEsse caso, estou supondo que é a nota de 5 alunos de uma turma maio, logo seria uma amostra.

```python
cvr = statistics.covariance(escrita,leitura)
print('Covariancia dos dois conjuntos é igual a =', cvr)
```

```python
plt.scatter(escrita,leitura)
```

```python
sns.regplot(x=escrita, y = leitura)
```

PAra valores muito elevados é possivel identificar uma relação de crescimento de um dependendo do outro, porem alguns valores baixos tem comportamentos distintos. Mesmo assim existe uma certa covariancia alta entre os 2 conjuntos
