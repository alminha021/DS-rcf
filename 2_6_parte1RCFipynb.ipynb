```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

importando tabela de valores passados no excel, emp = empregados e des = dsempregados, tudo em porcentagem e por data.
coluna 1 para idades, coluna 2 para valores de porcentagem de empregado e coluna 3 s√£o os valores de porcentagem para o desemprego

```python
data = pd.read_excel("dadosempregados.xlsx")
data.to_csv('dataset')
#df = pd.DataFrame(data, columns=['apple','alfa','boa'])
data.head(6)
```

utilizando o numpy e matplot para criar um grafico com colunas agrupadas, baseado na tabela com os valores de porcentagem de empregados e desempregados de acordo com a idade

```python
dados_empregados = data['emp']
dados_desempregados = data['des']

r1 = np.arange(len(dados_empregados))
largura = 0.25

r2 = [x + largura for x in r1]


plt.bar(r1,data['emp'], width=largura, label ="empregados")
plt.bar(r2,data['des'], width=largura,label ="desempregados")


plt.xticks([r +largura for r in range(len(dados_empregados))],['18 a 25','25 a 35','35 a 45','45 a 55','55 a 65','+65'])

plt.legend()
plt.show()
```
