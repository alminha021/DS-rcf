```python
import random
import string

class Crianca:
  def __init__(self):
    self.nome = ''
    for i in range (0, 10, 1):
        self.nome = self.nome + random.choice(string.ascii_uppercase)
    self.idade = random.randint(6,16)
    self.turno = 'M' if random.randint(0,1) == 0 else 'T'

def imprime(lista):
    for c in lista:
        print(c.nome, c.idade, c.turno, sep ='\t')
    print()

def organizaTurno(lista):
    #n = len(lista)
    aux = [None] * len(lista)
    index_manha = 0
    index_tarde = 0

    for crianca in lista:
        if crianca.turno == 'M':
            index_tarde += 1

    index_tarde_start = index_tarde
    for crianca in lista:
        if crianca.turno == 'M':
            aux[index_manha] = crianca
            index_manha += 1
        else:
            aux[index_tarde_start] = crianca
            index_tarde_start += 1
    print("ordenadas por turno: M e T")
    return aux

def organizaIdade(lista):
    def particionaFaixas(lista, inicio, fim):
        low = inicio
        high = fim
        mid = inicio

        while mid <= high:
            if 6 <= lista[mid].idade <= 10:
                lista[low], lista[mid] = lista[mid], lista[low]
                low += 1
                mid += 1
            elif 15 <= lista[mid].idade <= 16:
                lista[mid], lista[high] = lista[high], lista[mid]
                high -= 1
            else:
                mid += 1

    for h in range(len(lista)):
        if lista[h].turno == 'T':
            fim_manha = h
            break

    particionaFaixas(lista,0,fim_manha -1)
    particionaFaixas(lista,fim_manha,len(lista)-1)

    print("ordenadas range por partiçao: 6-10, 11-14, 15-16, mantendo o turno M e T")
    return lista

n = 20;
criancas = []
for i in range(1,21,1) :
    criancas.append(Crianca())

imprime(criancas)
aux = organizaTurno(criancas)
imprime(aux)
aux2 = organizaIdade(aux)
imprime(aux2)
```
